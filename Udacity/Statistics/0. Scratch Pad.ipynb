```python
x = np.array([2, -3, 5, 4, 7])
y = np.array([10, 13, 15, 10])

xbar = np.mean(x)
ybar = np.mean(y)

print('x-xbar =', x-xbar)
print('y-ybar =', y-ybar)

print('(x-xbar)^2 =', (x-xbar)**2)
print('(y-ybar)^2 =', (y-ybar)**2)

print('sum((x-xbar)^2) =', np.sum((x-xbar)**2))
print('sum((y-ybar)^2) =', np.sum((y-ybar)**2))

sp_squared = (np.sum((x-xbar)**2) + np.sum((y-ybar)**2))/(len(x)+len(y)-2)
print('pooled variance: ', sp_squared)

se = np.sqrt(sp_squared * (1/len(x) + 1/len(y)))
print('standard error:', se)

t = (xbar - ybar) / se
print('t =', t)
```
