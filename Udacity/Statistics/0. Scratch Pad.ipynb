```python
import numpy as np
import pandas as pd
from timeit import default_timer as timer
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
samples = np.random.binomial(15, p=0.9, size=10000)
counts = np.unique(samples, return_counts=True)
counts = (counts[0], counts[1]/sum(counts[1]))
plt.bar(counts[0], counts[1])
```

```python
samples = np.random.binomial(11, p=0.3, size=10000)
counts = np.unique(samples, return_counts=True)
counts = (counts[0], counts[1]/sum(counts[1]))
print(counts)
plt.bar(counts[0], counts[1])
```

```python
samples = np.random.binomial(10, p=0.8, size=10000)
counts = np.unique(samples, return_counts=True)
counts = (counts[0], counts[1]/sum(counts[1]))
print(counts[0], np.cumsum(counts[1]))
plt.bar(counts[0], counts[1])
```
