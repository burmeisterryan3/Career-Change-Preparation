```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read in the data
df = pd.read_csv('../data/pokemon.csv')
print(df.shape)
df.head()
```

### Example 1 - Demonstrate `figure.add_axes()` and `axes.hist()`

```python
# Create a new figure
fig = plt.figure()

# The argument add_axes represents the dimensions [left, bottom, width, height] of the new axes.
# All quantities are in fractions of figure width and height.
ax = fig.add_axes([.125, .125, .775, .755])
ax.hist(data=df, x='speed')
```

### Example 2 - Use `Axes` with `seaborn.countplot()`

```python
fig = plt.figure()
ax = fig.add_axes([.125, .125, .775, .755])
base_color = sns.color_palette()[0]
sns.countplot(data=df, x='generation_id', color=base_color, ax=ax)
```

### Example 3 - Sub-plots

```python
# Resize to a larger figure to allow for two plots side-by-side
plt.figure(figsize=[20, 5]) # width, height in inches

# histogram on left, example of too-large bin size
plt.subplot(1, 2, 1) # 1 row, 2 cols, subplot 1
bin_edges = np.arange(0, df['speed'].max()+4, 4)
plt.hist(data=df, x='speed', bins=bin_edges);

# histogram on right, example of too-small bin size
plt.subplot(1, 2, 2) # 1 row, 2 cols, subplot 2
bin_edges = np.arange(0, df['speed'].max()+1/4, 1/4)
plt.hist(data=df, x='speed', bins=bin_edges);
```

### Example 4 - Demonstrate `pyplot.sca()` and `pyplot.text()` to generate a grid of subplots

```python
fig, ax = plt.subplots(3, 4) # 3 rows, 4 cols (3x4 grid) of subplots
ax = ax.flatten() # reshape from 3x4 array into 12-element vector
for i in range(ax.shape[0]):
    plt.sca(ax[i]) # set the current Axes
    plt.text(0.5, 0.5, i+1) # print conventional subplot index number to middle of Axes; Axes limits are [0, 1] by default
```
