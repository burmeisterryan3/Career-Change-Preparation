```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../../../data/pokemon.csv')
print(df.shape)
df.head()
```

### Example 1 - Plot a default histogram using `matplotlib`

```python
# plt.hist(df['speed'])
plt.hist(data=df, x='speed')
```

### Example 2 - Histogram with fixed number of bins

```python
plt.hist(data=df, x='speed', bins=20)
```

### Example 3 - Histogram with dynamic number of bins (integer boundaries)

```python
# Create bins with step-size 5
# Add +5 to make sure all values are covered
step_size = 5
bins = np.arange(0, df['speed'].max()+step_size, step_size)
plt.hist(data=df, x='speed', bins=bins);
```

### Example 4 - Seaborn's `displot()`

`displot` is a figure-level function for histograms.

```python
# sns.displot(df['speed'])
sns.displot(data=df, x='speed')
```

### Example 5 - Seaborn's `displot` with `kde=True` and specified bins

Kernel density estimation is one way of estimating the probability density function of a variable. As a probability density function, the area under the curve is equal to 1.

The default settings in seaborn use a normal-distribution kernel. The probability of an outcome falling between any two values is equivalent to the area under the curve between the two points.

```python
# Plot with kde - kernel density estimation - https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot
sns.displot(df['speed'], kde=True, bins=bins)
```

```python
# Including additional parameters
sns.displot(df['speed'], kde=True, bins=bins, line_kws={'alpha': 1})
```

```python
# Removing the histogram
# Adjust bandwidth to show individual properties
sns.kdeplot(df['speed'], fill=True, bw_adjust=1);
```

### Example 6 - Seaborn's `histplot`

`histplot` is an `Axes`-level function for histograms.

```python
sns.histplot(df['speed'], kde=True, bins=bins)
```

### Example 7 - Plot two histograms side-by-side

This example puts two plots side by side through use of the [`subplot`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html) function, whose arguments specify the number of rows, columns, and index of the active subplot (in that order). The [`figure()`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html) function is called with the `figsize` parameter so that we can have a larger figure to support having multiple subplots.

```python
# Resize the chart, and have two plots side-by-side
# Set a larger figure size for subplots
plt.figure(figsize = [20, 5]) # 20 inch wide, 5 inch high

# histogram on left, example of too-large bin size
plt.subplot(1, 2, 1) # 1 row, 2 cols, subplot 1
bins = np.arange(0, df['speed'].max()+4, 4)
plt.hist(data=df, x='speed', bins=bins);

# histogram on right, example of too-small bin size
plt.subplot(1, 2, 2) # 1 row, 2 cols, subplot 2
bins = np.arange(0, df['speed'].max()+1/4, 1/4)
plt.hist(data=df, x='speed', bins=bins);
```
