```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../../data/fuel-econ.csv')
print(df.shape)
df.head()
```

### Example 1 - Jitter

```python
# Resize the figure to accommodate two plots
plt.figure(figsize = [20, 5])

# Plot the left plot: simple scatter plot
plt.subplot(1, 2, 1) # 1 row, 2 cols, subplot 1
sns.regplot(data = df, x = 'year', y = 'comb', truncate = False) # truncate = True - bound line to data limits; truncate = False - line extends to axes limits

# Plot the right plot: scatter plot with jitter
plt.subplot(1, 2, 2) # 1 row, 2 cols, subplot 2
sns.regplot(data = df, x = 'year', y = 'comb', truncate = False, x_jitter = 0.3); # x_jitter - add uniform random noise +/- 0.3 to x values
```

### Example 2 - Jitter and Transparency

##### `seaborn`

```python
# scatter_kws - dictionary of keyword arguments to be passed to the function used to draw the scatterplot
# alpha - transparency, [0, 1] where 0 is transparent and 1 is opaque
sns.regplot(data = df, x = 'year', y = 'comb', truncate = False, x_jitter = 0.3, scatter_kws = {'alpha' : 1/20});
```

##### `matplotlib`

```python
# Alternative approach
sns.regplot(data = df, x = 'year', y = 'comb', truncate = False, x_jitter = 0.3, scatter_kws = {'alpha' : 1/20});
# The scatter function below does NOT have a jitter parameter
plt.scatter(data = df, x = 'year', y = 'comb', alpha = 1/20);
```
