# Bivariate Visualizations

Bivariate visualizations are used to view the relationship between two variables. Often the following are used.

* Quantitative variable vs quantitative variable
    * Scatter Plot
    * Line Plot
    * Heat Map
* Quantitative variable vs qualitative variable
    * Box Plot
    * Violin Plot
    * Bar Chart
    * Point Plot
* Qualitative variable vs qualitative variable
    * Clustered Bar Chart

### Scatter Plot

Scatter plots are used to observe the relationship between two quantitative variables. Each data point is plotted individually as a point, its x-position corresponding to one feature and its y-position corresponding to the second.

The Pearson correlation coefficient, annotated $r$, can be used to quantify the strength of *linear* correlation between the two quantitative variables. Remember, positive $r$ values indicate a positive correlation (an increase in one is correlated to an increase in the other) while negative $r$ values indicate a negative correlation. The closer the value of $r$ is to $1$ or $-1$, the stronger the linear relationship. A value close to zero indicates the variables are uncorrelated.

To plot the scatter plots for all numeric variables in a dataset, `seaborn.PairGrid` can be used. This can be helpful to quickly view whether any of the variables are correlated.

See [scatterplots.ipynb](examples\Multivariate%20Exploration\Quantitative-Quantitative%20Plots\scatterplots.ipynb) for examples.

##### Overplotting, Transparency, and Jitter

If we have a very large number of points to plot or our numeric variables are discrete-valued, then it is possible that using a scatterplot straightforwardly will not be informative. The visualization will suffer from *overplotting*, where the high amount of overlap in points makes it difficult to see the actual relationship between the plotted variables. We can resolve this issue with sampling, transparency, and jitter.

By sampling, we can reduce the number of data points on our plot.
With transparency, overlapping data points will result in darker regions of our plots allowing us to see where the peaks in our data occur.

When dealing with discrete data, transparency will not help any overlapping data will be directly overlapping. In the case of discrete data, we can use jitter to slightly shift each point off of their discrete value. However, we should ensure that the jitter does not move a point into another discrete value, e.g, jitter for a discrete value of $1$ remains outside of the boundaries for $0$ and $2$.

See [overplotting-transparency-jitter.ipynb](examples\Multivariate%20Exploration\overplotting-transparency-jitter.ipynb) for examples.

### Heat Map

A heat map is a 2-d version of a histogram that can be used as an alternative to a scatterplot. Because we are using 2-D, there is not room for bar heights. Instead, counts are indicated by grid cell color.

* Heat maps are good for discrete variable vs discrete variable visualizations
* Heat maps are a good alternative to `transparency` for a lot of data.
* Like histograms, the choice of bin size affects the way we visualize and perceive our data.
    * With a bin size that is too large, you may miss trends in the data.
    * with a bin size that is too small, you may be distracted by noise.

One interesting way to use a heat map is to compute the correlation matrix for the numeric variables and plot the values using a heat map. This can be helpful to quickly view the correlation among all numeric variables.

### Violin Plots

In a violin plot, for each level of the categorical variable, a distribution of values on the numeric variable is plotted. The distribution is plotted as a kernel density estimate.

See [violin-plots.ipynb](examples/Multivariate%20Exploration/Categorical-Quantitative%20Plots/violin-plots.ipynb) for examples.

### Box Plots

A box plot (aka box-and-whisker plot) is another way of showing the relationship between a numeric variable and a categorical variable. Compared to the violin plot, the box plot leans more on the summarization of data by reporting a set of descriptive statistics for the numeric values on each categorical level.

The middle line is the *median* (i.e. *Q2*) value. The ends of the box are the first and third quartile values (i.e. *Q1* and *Q3*) while the minimum and maximum values (i.e., *Q0* and *Q4*) are drawn by the whiskers. $25\%$ of the values lie between each line.

Some box plots truncate their whisker lines and plot outliers separately. The most common truncation point is 1.5 times the interquartile range (IQR) above and below $Q3$ and $Q1$ respectively.
$$\text{IQR} = \text{Q3} - \text{Q1}$$

For example, if we have $Q1 = 53$ and $Q3 = 58.5$ with a maximum value of $68$. Then, $\text{IQR}=58.5-53=5.5$ and $1.5\times\text{IQR}=8.25$. The upper whisker bound is $58.5+8.25=66.75$ where the maximum would be drawn as an outlier as a point (or dot) above the whisker.

Note that the quartile ranges can be drawn on violin plots. However, for the sake of simplicity and focus, a box plot is recommended if attempting to communicate the quartiles.

### Stacked Bar Chart

The most basic stacked chart takes a single bar representing the full count, and divides it into colored segments based on frequencies on a categorical variable. If this sounds familiar, that's because it almost perfectly coincides with the description of a pie chart, except that the shape being divided is different.

Stacked bar charts and histograms are not uncommon, but there are often better plot choices available.

### Point Plots

Point plots allow for comparisons between with a categorical variable and second variable statistic (e.g., mean), similar to a bivariate bar chart. For instance, rather than viewing the mean as a bar, we can plot it as a point. This is particularly useful when having bars in reference to a 0 baseline isn't important or would be confusing.

### Line Plots

A line plot is used to show the trend of one numeric variable against values of a second variable. The line is meant to emphasize the sequential or connected nature of the x-values.

In contrast to a scatterplot, where all data points are plotted, in a line plot, only one point is plotted for every unique x-value or bin of x-values (like a histogram). If there are multiple observations in an x-bin, then the y-value of the point plotted will be a summary statistic (e.g., mean or median) of the data in the bin.

If the x-variable represents time, then a line-plot is frequently known as a tim series plot.

See [line-plots.ipynb](examples\Multivariate%20Exploration\Quantitative-Quantitative%20Plots\line-plots.ipynb) for examples.

### Q-Q Plots

Quantile-quantile (Q-Q) plots allow you to see how closely your quantitative data follows some hypothetical distribution. This is useful for certain parametric statistical tests, e.g., when checking for assumptions of normality. (Note: When testing for normality, you can use `scipy`'s [shapiro](https://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.stats.shapiro.html) function to perform the [Shapiro-Wilk Test](https://en.wikipedia.org/wiki/Shapiro%E2%80%93Wilk_test).)

The main idea for Q-Q plots is this: If the data was from a particular distribution (e.g., normal), then we'd expect a pattern for how far each data point is from the mean of the distribution. If we order our data points from smallest to largest, we could then compare how large the $k\text{-th}$ ranked data point is against the $k\text{-th}$ point from the expected distribution.

To get the expected values, we can use `scipy`'s `norm` and `ppf` (percent point function) functions. The *ppf* takes a proportion ($[0, 1]$) and returns the value in the distribution that would leave that proportion of the curve to the left.

For a standard normal distribution (i.e., $\mu=0$ and $\sigma=1$), $ppf(0.25)=-0.674$, $ppf(0.5)=0$, $ppf(.075)=0.674$.

The question then becomes what values to stick into the $ppf$. There are a few different conventions, but they generally take the form. Given $n$ data points the $k\text{-th}$ value should be at probability point $\frac{k-a}{n+1-2a}$, for some $a$ between $[0, 1]$.

This equation distributes the probability points symmetrically about $0.5$, and adjusting $a$ changes how much probability is left in the tails of the $[0, 1]$ range. Commonly, $a$ is set to a balanced value of $0.5$, which yields $\frac{k-0.5}{n}$.

### Faceting

In faceting, the data is divided into disjoint subsets, most often by different levels of a categorical variable. For each of these subsets of the data, the same plot type is rendered on other variables. Faceting is a way of comparing distributions or relationships across levels of additional variables, especially when there are three or more variables of interest.

# Multivariate Exploration of Data

When attempting to visualize three or more variables, visualizing our data in 2-D becomes much more challenging. Typically, we use encodings to represent extra dimensions or variables in our data.

### A Cautionary Tale

When you have only two variables plotted, A and B, you have only one relationship to understand. Add in a third variable C, and you have three pairwise relationships: A vs. B, A vs. C, and B vs. C. You also need to consider one interaction effect between all three variables: Does the value of C affect the relationship between A and B? With a fourth variable, you have six possible pairwise relationships and four different three-variable interaction effects. This exponential explosion of possible relationships with the number of variables means that there is a potential for cognitive overload if the data isn't conveyed clearly.

When you move on to explanatory data visualizations, try to limit the number of variables that are introduced at the same time and make sure that the encoding choices convey the main findings to your reader in the clearest way possible. While it's good to keep a soft limit of about three or four variables in a single visualization, you can exceed this if the trends are clear or you introduce features to your reader in a systematic way.

### Non-positional Encodings for Third Variables

There are four major cases to consider when we want to plot three variables together:
* three numeric variables
* two numeric variables and one categorical variable
* one numeric variable and two categorical variables
* three categorical variables

If we have at least two numeric variables, as in the first two cases, one common method for depicting the data is by using a scatterplot to encode two of the numeric variables, then using a non-positional encoding on the points to convey the value on the third variable, whether numeric or categorical.

The three main non-positional encodings are *shape*, *size*, and *color*.

##### Shape

Shape is a good encoding for categorical variables, using one shape for each level of the categorical variable.

##### Size

Point size is a good encoding for quantitative variables. Generally, we want the numeric values to be proportional ot the area of the point markers.

##### Color

Color is a very common encoding for either qualitative or quantitative variables. The category in a clustered bar chart or count in a heat map are two examples from [Lesson 4](4.%20Bivariate%20Exploration%20of%20Data.ipynb).

See [scatterplots-non-positional-encodings.ipynb](examples\Multivariate%20Exploration\Quantitative-Quantitative%20Plots\scatterplots-non-positional-encodings.ipynb) for examples.

### Color Palettes

There are three major classes of color palettes to consider: *qualitative*, *sequential*, and *diverging*.

##### Qualitative

Qualitative palettes are built for nominal-type data. Consecutive color values are distinct so that no inherent ordering of levels is implied. Colors in a good qualitative palette should also try and avoid drastic changes in brightness and saturation that would cause a reader to interpret on category as being more important that the others.

Example palettes: 'deep', 'pastel', 'dark', 'muted', 'bright', 'colorblind'

##### Sequential

Sequential palettes are good for ordered categorical variables with few levels. Typically, light shading indicates smaller values while darker colors indicate larger values, though it can be reversed.

Example palettes: 'rocket' (white-orange-red-purple-black), 'mako' (mint-green-blue-purple-black)

##### Diverging

Diverging palettes are used when there is a meaningful center or zero for our variable. In a diverging palette, two sequential palettes with different hues are put back to back, with a common color (usually white or gray) connecting them. One hue indicates values greater than the center point, while the other indicates values smaller than the center.

Example palettes: 'vlag' (blue-white-red), 'icefire' (blue-black-orange)

##### Warnings

1. Be aware of and consider color blindness when selecting colors.
1. Be aware of the effect of transparency and overlap on interpretability. If points of different color on a qualitative scale overlap, they may result in a third color that cannot be matched to something in the palette. If multiple points on a quantitative scale overlap, then the result may be a value that does not actually exist in the data. You should avoid or minimize transparency in plots with a color.

### Feature Engineering

As you explore your dataset, you might find that two variables are related in some way. Through feature engineering, we can create new variables by performing a mathematical operation (e.g., add, subtract, multiply, or divide) on the original variables that provides deeper insight into our data.

For instance, the total number of crimes in a region may be interesting, but normalizing the data by population to obtain the crime *rate* might suggest that crime is more likely to occur by chance simply due to population increases.

Another way that you can perform feature engineering is to use the `pandas.cut` function to divide a numeric variable into ordered bins. When we split a numeric variable into ordinal bins, it opens it up to more visual encodings. For example, we might facet plots by bins of a numeric variable, or use discrete color bins rather than a continuous color scale. This kind of discretization step might help in storytelling by clearing up noise, allowing the reader to concentrate on major trends in the data. Of course, the bins might also mislead if they're spaced improperly â€“ check out [this PolicyViz post](https://policyviz.com/2017/11/02/choosing-map-bins/) if you'd like to see a deeper discussion in the context of map-based visualizations.

### Example

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the data
df = pd.read_csv('data/fuel-econ.csv')
```

```python
# Convert the 'VClass' column from a plain object type into an ordered categorical type
vclasses = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(ordered=True, categories=vclasses)
df['VClass'] = df['VClass'].astype(vclasses)
```

```python
plt.figure(figsize=(20, 5))
base_color = sns.color_palette()[0]

# Left plot: violin plot
plt.subplot(1, 3, 1)
sns.violinplot(data=df, x='VClass', y='comb', color=base_color, inner='quartile')
plt.xticks(rotation=45)
plt.xlabel('Vehicle Class')
plt.ylabel('Combined Fuel Eff. (mpg)')

# Center plot: box plot
plt.subplot(1, 3, 2)
sns.boxplot(data=df, x='VClass', y='comb', color=base_color)
plt.xticks(rotation=45)
plt.xlabel('Vehicle Class')
plt.ylabel('Combined Fuel Eff. (mpg)')

# Right plot: adapted bar chart
plt.subplot(1, 3, 3)
sns.barplot(data=df, x='VClass', y='comb', color=base_color)
plt.xticks(rotation=45)
plt.xlabel('Vehicle Class')
plt.ylabel('Avg Combined Fuel Eff. (mpg)');
```
