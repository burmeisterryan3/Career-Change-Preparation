### Motivating Data Visualization

Summary statistics like the mean and standard deviation can be great for attempting to quickly understand aspects of a dataset, but they can also be misleading if you make too many assumptions about how the data distribution looks.

##### Example - Anscombe's Quartet

Here are four different sets of a data with the same mean and standard deviations for both $X$ and $Y$ values. Though their descriptive statistics are the same, note the difference in their plots.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./data/anscombes-quartet.csv')

fig, ax = plt.subplots(2, 2, figsize=(10, 10))
xlims = [[0, 20]]*4
ylims = [[0, 13]]*4

for i, (x, y, xlim, ylim) in enumerate(zip(['X1', 'X2', 'X3', 'X4'], ['Y1', 'Y2', 'Y3', 'Y4'], xlims, ylims)):
    ax[int(i/2), i%2].scatter(x=df[x], y=df[y])
    ax[int(i/2), i%2].set_xlim(0, 20)
    ax[int(i/2), i%2].set_ylim(0, 13)
    ax[int(i/2), i%2].set_xlabel(x)
    ax[int(i/2), i%2].set_ylabel(y)
    ax[int(i/2), i%2].set_xticks(np.arange(0, 21, 5))
    ax[int(i/2), i%2].set_yticks(np.arange(0, 15, 2))
    ax[int(i/2), i%2].text(0.025, 0.9, "Mean: {}\nStd: {}".format(round(df[x].var(), 2), round(df[x].std(), 2)),
                           bbox=dict(facecolor='white', alpha=0.5, boxstyle='round'), transform=ax[int(i/2), i%2].transAxes)
fig.suptitle('Anscombe\'s Quartet', fontsize=16)
plt.show()
```

```python
sns.scatterplot(x='X1', y='Y1', data=df)
sns.scatterplot(x='X2', y='Y2', data=df)
sns.scatterplot(x='X3', y='Y3', data=df)
sns.scatterplot(x='X4', y='Y4', data=df)
```

### Exploratory vs Explanatory Analyses

* Exploratory Analysis
    * Performed when you are searching for insights.
    * You are the consumer of the plots and are looking for answers to your own questions. They do not need to be aesthetically pleasing.
* Explanatory Analysis
    * Performed when you are providing results for others.
    * Visualizations should be accurate, insightful, and visually appealing while providing emphasis to the message you are attempting to convey.

##### Five Steps for Data Analysis

* Extract
    * Obtain data.
* Clean
    * Possible use of exploratory visuals
* Explore
    * Exploratory visuals
* Analyze
    * Explanatory or exploratory visuals
* Share
    * Explanatory visuals

### Visualization Libraries

Three popular visualization libraries include:

* [Matplotlib](https://matplotlib.org/)
    * Offers the greatest flexibility, but it can take some effort to put together common visualizations.
* [Seaborn](https://seaborn.pydata.org/)
    * Built on top of matplotlib and adds functions to make common statistical visualizations easier to generate.
* [Pandas](https://pandas.pydata.org/)
    * Provides some convenient tools for visualizing data that hooks into matplotlib.
