```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('results.csv')
df.head()
```

```python
fig = plt.figure(figsize = [10, 10])
fig.suptitle('Performance of Different Architectures', fontsize=16)
fig.text(0.5, 0.04, 'Architecture', ha='center', fontsize=14)
fig.text(0.04, 0.5, 'Percentage Correct', va='center', rotation='vertical', fontsize=14)

default_color = sns.color_palette()[0]

plt.subplot(2, 2, 1) # 2 rows, 2 columns, 1st subplot
sns.barplot(x='arch', y='perc-not-dog-correct', data=df, color=default_color)
plt.title('\"Not-a-Dog\" Pictures Correctly Classified')
plt.xlabel('')
plt.ylabel('')
plt.ylim(0, 100)

plt.subplot(2, 2, 2) # 2 rows, 2 columns, 2nd subplot
sns.barplot(x='arch', y='perc-dog-correct', data=df, color=default_color)
plt.title('Dog Pictures Correctly Classified')
plt.xlabel('')
plt.ylabel('')
plt.ylim(0, 100)

plt.subplot(2, 2, 3) # 2 rows, 2 columns, 3rd subplot
sns.barplot(x='arch', y='perc-breed-correct', data=df, color=default_color)
plt.title('Breeds Correctly Classified')
plt.xlabel('')
plt.ylabel('')
plt.ylim(0, 100)

plt.subplot(2, 2, 4) # 2 rows, 2 columns, 4th subplot (bottom right)
sns.barplot(x='arch', y='perc-match-label', data=df, color=default_color)
plt.title('Overall Accuracy')
plt.xlabel('')
plt.ylabel('')
plt.ylim(0, 100)

plt.savefig('results.png')
```
