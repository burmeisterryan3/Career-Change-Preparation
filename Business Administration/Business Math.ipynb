## Basic Math

* Simple Interest
    * An interest charge that borrowers bay lenders for a loan using only the principal cost without compounding interest
    * Given a principal amount $P$, an interest rate $r$, and a loan term of $n$ years, then
        * Simple Interest = $P \times r \times n$
* Compounding Interest with Fixed Rate
    * Given a rate, $r$, and an initial value of $P_1$, then the value after a single period is
        * $P_2 = P_1 + (r\times P_1)$
    * Given a fixed rate, $r$ and an initial value of $P_1$, then the value at time $t$, $P_t$, is
        * $P_t = P_1\times(1+r)^t$
* Weighted Average
    * For a set of numbers $P_1$, $P_2$, $\dots$, $P_n$ with corresponding weights $W_1$, $W_2$, $\dots$, $W_n$, then the weighted average is $\frac{\sum_i W_i\times P_i}{\sum_j W_j}$

## Business Math

### Abbreviations

* COGS - Cost of Goods Sold
* EBITDA - Earnings Before Interest, Taxes, Depreciation, and Amortization
* R&D - Research & Development
* ROI - Return on Investment
* SG&A - Selling, General, & Administrative

### Income Statement

* $\text{Gross Profit} = \text{Revenue} - \text{COGS}$
    * Revenue = Sales
* $\text{Gross Margin} = \frac{\text{Gross Profit}}{\text{Revenue}}$
* $\text{Other Expenses} = \text{SG\&A} + \text{R\&D} + \text{Depreciation \& Amortization} + \text{Interest}$
* $\text{Earnings Before Tax} = \text{Gross Profit} - \text{Other Expenses}$
* $\text{Taxes} = \text{Earnings Before Tax} \times \text{Tax Rate}$
* $\text{Net Income} = \text{Earnings Before Tax} - \text{Taxes}$
    * Net Income = Earnings

### Investments

* Breakeven
    * Revenue = Costs (fixed and variable)
    * $\text{Price} \times \text{Volume} = \text{Fixed Costs} + (\text{Variable Costs} \times \text{Volume})$
    * Assuming production costs per unit = $35,000 and a sale price per unit of $100,000, then
        * $100,000 * Volume = $800,000 + ($35,000 * Volume)
            * NOTE: $800,000 is the "Total Other Expenses" or "Fixed Cost" below
        * Volume = $800,000 / $65,000 = 12.3 (units to be sold to breakeven)
* ROI
    * Given an investment, a time period's percentage of that investment (e.g., year)
    * $\text{ROI} = \frac{\text{Profit (or loss) from an Investment}}{\text{Cost of Investment}}$
    * Assuming a $7,000,000 investment
        * ROI = $350,000 / $7,000,000 = 0.05 = 5%
* Payback Period
    * Time it takes to recoup investment cost
        * NOTE: Assumes no compounding effects, i.e., linear growth effects
    * $\text{Payback Period} = \frac{\text{Amount Invested}}{\text{Annual Return}}$
    * Assuming a $7,000,000 investment and profit of $350,000
    * Payback period = $7,000,000 / $350,000 = 20 years
* Rule of 72
    * $\text{Time to double investment} = \frac{72}{\text{Annual Rate of Return}}$
        * NOTE: Assumes compounding effects, i.e., exponential growth effects
    * Assuming a $7,000,000 investment with a 5% annual return rate
        * Time to double investment = 72 / 5 = ~14 years

### Valuation

* Market Capitalization
    * Value of a companies shares of stock
    * $\text{Market Capitalization} = \text{Total number of shares} \times \text{Market price of one share}$
    * $\text{Market Capitalization} = \frac{\text{P}}{\text{E}}\times\text{Net Income}$
    * Assuming 100,000 shares at $180 per share
        * Market Capitalization = $100,000 * $180 = $18,000,000
* Enterprise Value
    * Value of a company as a whole, i.e., price to pay to purchase all shares of the company and pay its net debt
    * $\text{Enterprise value (EV)} = \text{Market Capitalization} + \text{Total Debt (short and long term)} - \text{Cash and equivalents}$
    * $\text{EV} = \frac{\text{EV}}{\text{S}}\times\text{Revenue}$
    * Assuming 100,000 shares at $180 per share, $0.5m of debt, and $4.2m in cash
        * EV = $18,000,000 + $500,000 - $4,200,000 = $14,300,000
* Comparable Analysis
    * The value of a business is determined by applying valuation ratios of comparable businesses
    * Common valuation ratios
        * Revenue Multiples - Enterprise Value to Sales (EV/S), Price to Sales (P/S)
        * Profit Multiples - Price to Earning (P/E), Enterprise Value to Earnings (EV/E)
    * Assuming $0.5m of debt, $4.2m of cash, $3m of current year sales, and a net income of $0.875m while using the below table
        * Using the EV/S method
            * EV = 9 * $3m = $27m
            * Market Capitalization = EV - Debt + Cash = $27m - $0.5m + $4.2m = $30.7m
        * Using the P/E method
            * Market Capitalization = P/E * Earnings = 32 * $0.875m = $28m
* Discounted Cash Flow (DCF) Valuation
    * EV = Net Present Value of Future Cash Flows
    * Concepts to Understand
        * Time Value of Money
        * Discount Rate
        * Net Present Value
        * Terminal Value
    * Time Value of Money
        * $\$1000$ today will be worth $\$1000\times(1+i)$ in a year, where $i$ is the interest rate expected
        * Thus, the present value (value today) of receiving $\$1,000$ in a year, $\frac{\$1,000}{1+i}$
            * NOTE: This means for any investment with a positive return, the value of the same amount today is worth more than the same amount later
    * Discount Rate
        * Discount rate is the minimum possible return rate that will make an investment worthwhile for an investor
        * Alternatively said, it is the opportunity cost of putting the cash in an investment carrying similar risk
            * The riskier the investment the higher the discount rate
            * Example: The more volatile the industry and the more debt a company is carrying means greater risk, and therefore, a greater discount rate
        * Example: Investors require 1% interest when lending to a government.
            * The present value of receiving $1,000 from a government in one year would be $1,000 / (1 + 1%) = $990
        * Example: Investors require 20% interest when lending to a start-up.
            * The present value of receiving $1,000 from that start-up in a year would be $1,000 / (1 + 20%) = $833
    * Net Present Value (NPV)
        * The NPV of cash flows is the sum of the Present Value of cash flows over a period
        * $\text{NPV} = \sum\frac{\text{CF}_{\text{t}}}{(1+\text{r})^{\text{t}}}$
            * $\text{CF}_{\text{t}}$ - net cash flow during a single period $\text{t}$
            * $\text{r}$ - discount rate
        * Assuming we are in year 0 and the cash flow in year 1 and 2 are $0.8m and $1.5m respectively with a 5% discount rate
            * NPV = $\frac{.8}{(1+5\%)^1}$ + $\frac{1.5}{(1+5\%)^2}$ = $0.76m + $1.36m = $2.12m
    * Terminal Value (TV)
        * The TV is the value of the asset beyond the period when future cash flows have been estimated
        * $\text{TV} = \frac{\text{CF}_{\text{t}}}{\text{r}}$
            * $\text{CF}_{\text{t}}$ - net cash flow during a single period $\text{t}$
            * $\text{r}$ - discount rate
                * NOTE: assumes cash flow stays constant over time
        * Assuming $1.5m of steady cash flows in perpetuity
            * TV = $\frac{1.5}{5\%}$ = $30m
    * Enterprise Value
        * $\text{EV} = \text{NPV} + \frac{\text{TV}}{(1+\text{r})^{\text{t}}} = \sum\frac{\text{CF}_{\text{t}}}{(1+\text{r})^{\text{t}}} + \frac{\text{CF}_{\text{t}}}{\text{r}(1+\text{r})^{\text{t}}}$
        * Assuming the values computed above
            * EV = 2.12 + $\frac{30}{(1+5\%)^2}$ = $2.12m + $27.21m = ~$29m

### Operations

* Capacity
    * The maximum level of output that a process can sustain
* Utilization
    * The amount of capacity actually being used for production over a set interval
    * $\text{Utilization} = \frac{\text{Quantity Produced}}{\text{Capacity}}$
    * Assuming we have a capacity of 100 over a period of time and produced 20
        * Utilization = 20 / 100 = .2 or 20%
* Process Cycle Efficiency
    * The amount of value-adding time in the lead time of a process
    * $\text{Process Cycle Efficiency} = \frac{\text{Value-Adding Time}}{\text{Lead Time}}$
        * Lead Time - Time from order to delivery
        * Value-Adding Time - Time spent on activities that are directly spent on production and distribution
    * Assuming it takes 30 days to produce and delivery a product and that 12 days are actually spent on production and distribution
        * Process Cycle Efficiency = 12 / 30 = .4 or 40%
