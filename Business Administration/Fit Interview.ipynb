## Fit Interview

A fit interview is aimed at assessing the personal attributes, presence, and communication abilities of a candidate. It seeks to answer whether the candidate can get things done and whether they can represent the firm.

### Ordering

1. Present Background
2. Discuss Achievements
    1. Should be tailored to the specific role for which you are applying
3. Share Motivations
    1. Why this job, e.g, consulting?
    2. Why this company or firm?

### Assessment

1. Rapport-Building
    1. Make a personal connection
    2. Remain engaging with a back-and-forth dialogue
    3. Show your unique personality traits
2. Performance Track Record
    1. Provide evidence of being a top-performer without boasting
3. Transferable Abilities
    1. Persuasion
    2. Teamwork
    3. Entrepreneurial drive and leadership
    4. Adaptability and learning orientation
    5. Working independently and under pressure
4. Motivation
    1. Needs to clearly articulate why this role, company, and time

### Narrative Arc

1. Situation
    1. What was your role and the problem before it?
2. Complications
    1. Without a challenge or complication, the result will seem less impressive.
3. Actions
    1. What did you do? Why did you do it?
    2. Ensure you speak in the 1st person.
4. Results
    1. Ensure your story has a happy ending.


### Influence Tactics

When describing your experiences, note that most firms will be looking assessing your transferable abilities. You do not need to specifically call these out during you narratives, but they should be self-evident as you walk through past performances, projects, or challenges.

Focus on the rational, social, and emotional tactics your brought to bear on the poblem before you.

* Rational
    * Logical Persuading - Using logic to explain what you believe or what you want
    * Legitimizing - Appealing to authority
    * Exchanging - Trading for cooperation - preferably implicitly
    * Stating - Asserting what you believe or want
* Social
    * Socializing - Getting to know the other person; being friendly; finding common ground
    * Appealing to Relationship - Securing the cooperation of people you already know well
    * Consulting - Involving people in the problem or solution or building alliances
    * Alliance-building - Finding supporters or building alliances
* Emotional
    * Appealing to values - Making an appeal to the heart
    * Modelling - Behaving in ways you want others to behave

* Dark-side tactics (Do not use these!)
    * Avoiding - Forcing others to act by avoiding responsibility or conflict
    * Manipulating - Influencing through lies, deceit, hoaxes, swindles, and cons
    * Intimidating - Forcing people to comply by being loud, overbearing, bullying, etc.
    * Threatening - Harming others or threatening to harm them if they do not comply
